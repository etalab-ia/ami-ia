```python
# lire le jeu de données en un objet "pandas dataframe"
path = '/Users/kimmontalibet/Desktop/datagouv/'
df = pd.read_csv(path + "completez-le-questionnaire-en-quelques-minutes.csv", sep = ",", header = 0)
#df = pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/1b33488b-3bf7-4054-947b-d3f301b2ee84", sep = ",", header = 0)
```

```python
# on renomme les colonnes des questions et on conserve un dictionnaire avec la correspondance 
colnames = list(df.columns)
dict_questions = dict(zip(["question_" + str(x) for x in range(1, 17)], colnames[6:23]))
colnames_new = colnames[0:6] + ["question_" + str(x) for x in range(1, 17)] + ["age"]
df.columns = colnames_new
```

```python
dict_questions
```

```python
df.head()
```

```python
# on regarde le nombre de réponses remplies pour chaque colonnes 

colonnes_questions = list(dict_questions.keys())

nb_reponses = df[colonnes_questions].describe().transpose()
nb_reponses["proportion_questions_remplies"] = nb_reponses["count"]/len(df)
nb_reponses.sort_values(by = "count", ascending = False)
```

```python
# remarque: seule la question 3 est à valeur unique 
print(dict_questions["question_3"])
print("\n")
question_3 = pd.DataFrame(df["question_3"].value_counts())
question_3["pourcentage"]  = question_3["question_3"] / sum(question_3.values)
question_3
```

```python
# on regarde le nombre de réponses manquantes par répondant
df["nb_reponses_manquantes"] = df.apply(lambda row: sum([pd.isnull(row[x]) for x in colonnes_questions ]), axis = 1)
```

```python
df["nb_reponses_manquantes"].describe()

# en moyenne 1 réponse manquante par personne, max à 16
# on peut éliminer les lignes avec 16/16 réponses manquantes
```

```python
df["nb_reponses_manquantes"].value_counts()
```

```python
df = df[df["nb_reponses_manquantes"]<16] 
```

```python
# nombre de caractères et de mots par réponse

for col in colonnes_questions: 
    df[col + "_nb_car"] = df[col].map(lambda x: len(x) if pd.isnull(x)==False else 0)
    
```

```python
colonnes_nb_car = [x + "_nb_car" for x in colonnes_questions]
df[colonnes_nb_car].describe().transpose().sort_values(by = "mean", ascending = False)
```
